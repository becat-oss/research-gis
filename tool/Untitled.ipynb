```python
import pandas as pd
import json
import jsonify
```

```python
def removeNumberFromIndex(df):
    def removeNumber(ix):
        return ix.split(' ')[1]
    return df.set_index(df.index.map(removeNumber))
```

```python
df1=removeNumberFromIndex(df)
```

```python
def formatDf(df,cellType):
    def removeNumberFromIndex(ix):
        return ix.split(' ')[1]
    
    def removeComma(cell):
        return cell.replace(',', '')
    
    df1=df.applymap(removeComma).astype(cellType)
    return df1.set_index(df1.index.map(removeNumberFromIndex))
    
```

```python
df1=formatDf(df,'int64')
```

```python
def exportToJson(df,col):
    jsonStr = df[col].to_json(force_ascii=False)
    return json.loads(jsonStr)
    #dictData=df[col].to_dict()
    #return json.loads(dictData)
```

```python
exportedJson=exportToJson(df1,'令和元年度')
```

```python
def writeJson(data,fileName):
    with open('dist/'+fileName+'.json', 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False)
```

```python
writeJson(exportedJson,'gdpDeltaPerCapita')
```
